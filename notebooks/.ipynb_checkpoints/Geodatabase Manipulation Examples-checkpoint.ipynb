# Geodatabase Manipulation Examples

This notebook contains examples for how to connect to, create and manipulate geodatabases using the `envelopegis.models` API.

## Importing the necessary components

For this example we will use the `Workspace()` and `Layer()` classes from the Models API for most of the work. We will also import the `Polygon()` class to use when creating new data.

```python
from envelopegis.models.workspace import Workspace
from envelopegis.models.layer import Layer
from envelopegis.models.polygon import Polygon
from envelopegis.models.point import Point
```

## Connecting to an existing geodatabase

For our example we'll copy some data to work with from an existing database. In the code below we will connect to a database called "sample_data" and print the list of layers available in it.

Notice that in this example we create an instance of a workspace object directly by passing the connection credentials in via the object constructor.

```python
# credentials for connecting to sample data
sample_credentials = {
    'host': 'gisbox.ehq',
    'port': 7601,
    'database': 'sample_data',
    'user': 'postgres',
    'password': ''
}

# connect to the samples database
sample_workspace = Workspace(sample_credentials)

# print the list of layers (a.k.a. feature classes) available in the workspace
print('Layers available in this geodatabase are:')
for layer_name in sample_workspace.list_feature_classes():
    print('   {}'.format(layer_name))
```

## Creating a new geodatabase

For this example we will create a temporary geodatabase and copy some data into the database to work with. Later in the example we will delete this temporary database to keep our database instance clean.

We'll use the a static method on the workspace class to create a new geodatabase for our example data. The object returned after creating the geodatabase is a workspace object which maintains a connection to the new geodatabase. In the code below, notice that no database is specified in the connection credentials. This is because we are connecting to the database instance in order to create a new geodatabase.

```python
# database connection credentials
credentials = {
    'host': 'gisbox.ehq',
    'port': 7602,
    'user': 'postgres',
    'password': ''
}

# create a geodatabase called "geodb_example" using the static `create_geodatabase` method from the workspace class
example_workspace = Workspace.create_geodatabase(credentials, 'geodb_example')
```

## Selecting and Copying Data

Now let's load up our sample feature classes and copy them to our new geodatabase. But first, we'll filter the data so that only a small selection of the datasets are copied (this will keep our working datasets small and fast).

### Selecting Data with Attributes

First, we'll select the lots in the `primary_lots` table that are on block "NY-MN-1392" and copy those features to the new geodatabase in a layer called `sample_lots`. Notice that in the query the string value is wrapped in single-quotes. This is only required for string values - numeric values can be passes in without quotes.

```python
# load up the `primary_lots` layer by instantiating a Layer object. The arguments 
# for the constructor are the name of the layer and the workspace where it resides
primary_lots_layer = Layer('primary_lots', sample_workspace)

# select the lots that are on block NY-MN-1392
primary_lots_layer.select_layer_by_attribute("CBB = 'NY-MN-1392'")

# copy the data to the example geodatabase by providing the workspace for the
# destination in the `output_workspace` parameter. Notice that calling
# the copy function only copies the selected data, not the entire dataset.
# The returned object is another `Layer()` object representing the newly copied layer.
example_lots_layer = primary_lots_layer.copy_to_new_table('example_lots', 
                                                         output_workspace=example_workspace)
```

### Selecting Data by Location

Now that we have a block's worth of data in our sample database, let's copy some zoning districts to work with. In this case we only want the zoning districts that intersect our sample block, so we will use a spatial query to find these zoning districts.

```python
# load up the `zoning_districts_aligned` layer from our samples database.
zoning_districts_layer = Layer('zoning_districts_aligned', sample_workspace)

# Select the zoning districts that intersect our selected lots 
zoning_districts_layer.select_layer_by_location('INTERSECT', primary_lots_layer)

# Copy the selected zoning districts to our example database
example_zd_layer = zoning_districts_layer.copy_to_new_table('example_zds',
                                                           output_workspace=example_workspace)
```

## Examing the `Layer()` object

The `Layer()` object is one of the most refined objects in the Models API, and has several useful properties and methods. In the code below we'll examine some of the useful properties of the `Layer()` object.

```python
# print some useful properties of the `zoning_districts_aligned` layer
print('Table name (in the geodatabase): {}'.format(example_zd_layer.table_name))
print('Row count: {}'.format(example_zd_layer.count))
print('Spatial reference: {}'.format(example_zd_layer.spatial_reference))
print('Id field name: {}'.format(example_zd_layer.oid))
print('Geometry field name: {}'.format(example_zd_layer.shape))
print('Fields:')
for field in example_zd_layer.fields:
    print('\t{} ({})'.format(field.name, field.type))
```

## Plotting the Data

Layer objects can be diplayed within a notebook by calling the `plot()` method. The `plot()` method uses the `matplotlib` library to render geometries and accepts most standard `matplotlib` inputs as parameters. 

The `plot()` method can be called with no arguments to render the layer in it's own figure, or a `matplotlib.axes` object can be passed to the `plot()` method to render the layer into an existing axes. The `plot()` method returns the `matplotlib.axes` object that the layer has been rendered into so that other geometries can be rendered into the same axes.

```python
# Plot the geometries in the layer
# and give the plot a title
ax = example_lots_layer.plot(color='0.8', edgecolor='0.4')
ax.set_title('Example lots layer')

# repeat the process for the zoning districts layer
ax = example_zd_layer.plot(color='White', edgecolor='Blue')
ax.set_title('Example zoning districts layer')

# now draw both layers in the same plot
# (note that we're using the boundary of the 
# zoning district geometry to create the 
# illusion of transparency)
ax = example_lots_layer.plot(color='0.8', edgecolor='0.4')
example_zd_layer.plot(boundary=True, color='Blue', ax=ax)
ax.set_title('Both example datasets')
```

## Creating layers

Workspace objects can be used to create new layers from scratch, and methods from the Layer object can then be used to create fields, insert, update, and delete rows, etc.

```python
# use the workspace object to create a new empty feature class

# the first parameter indicates the table name for the new table in the geodatabase,
# the second parameter indicates that the layer should hold polygons, and the
# third parameter indicates that the spatial reference for the data in the 
# dataset should be 2263 (New York State Plane)

# the function returns a `Layer()` object that can be used to further manipulate the layer
test_layer = example_workspace.create_feature_class('test_layer', 'POLYGON', spatial_reference=2263)

# by default, the new layer has two fields, "OBJECTID" and "SHAPE"
print('Default fields in new layer:')
for field in test_layer.fields:
    print('\t{} ({})'.format(field.name, field.type))
    
# let's add a new field to the layer called "code" which can hold text
test_layer.add_field('code', 'TEXT', field_length=10)

print('After adding the new field, fields are:')
for field in test_layer.fields:
    print('\t{} ({})'.format(field.name, field.type))
```

## Adding data to layers

Adding data to layers involves creating a list of dictionaries where each dictionary represents a row. The key/value pairs in each dictionary represent the field names and values for each row, and the list of dictionaries represents the collection of rows to be inserted into the layer.

In the example below, we will create a few dummy polygons to insert into the new layer as rows. We'll assign each polygon a 'code' value for labeling.

```python
# first, create a few polygons

# notice that the spatial reference of the layer is
# provided when creating the geometries

left_poly = Polygon([
    {'x': 0, 'y': 0},
    {'x': 10, 'y': 0},
    {'x': 10, 'y': 10},
    {'x': 0, 'y': 10},
], spatial_reference=test_layer.spatial_reference)

right_poly = Polygon([
    {'x': 15, 'y': 0},
    {'x': 25, 'y': 0},
    {'x': 25, 'y': 10},
    {'x': 15, 'y': 10},
], spatial_reference=test_layer.spatial_reference)

# now we'll create a list with two dictionaries, 
# representing 2 new rows for the layer
new_rows = [
    {
        'code': 'Left',
        test_layer.shape: left_poly
    },
    {
        'code': 'Right',
        test_layer.shape: right_poly
    },
]

# add the rows to the layer
test_layer.add_rows(new_rows)

# plot the data
ax = test_layer.plot(color='0.8', edgecolor='0.4')
test_layer.label('code')
ax.set_title('Number of rows: {}'.format(test_layer.count))
```

## Updating data in layers

Updating existing rows in a layer is performed using an update cursor, similar to the search cursor used above. Both search and update cursors only operate on selected rows when the layer contains a selection set. If the layer does not contain a selection the cursors will operate on all rows in the table.

```python
# for the sake of example, let's switch the code values
# to the opposite of what they should be using an update cursor

with test_layer.get_update_cursor(['code']) as cursor:
    for row in cursor:
        row['code'] = 'Right?' if row['code'] == 'Left' else 'Left?'
        cursor.update_row(row)

# plot the data
ax = test_layer.plot(color='0.8', edgecolor='0.4')
test_layer.label('code')
ax.set_title('Number of rows: {}'.format(test_layer.count))
```

## Updating data using the `calculate_field()` method

Another way to update data is by using the `calculate_field()` method, which will set a field to a specific value for all selected rows (or all row if there is not a selection on the table). The calculate field method is useful when you want to apply the same value to field for multiple rows or for the entire table.

```python
# let's use the `calculate_field()` method to set
# the 'code' attribute in our test layer to "Square"
# for all rows in the table
test_layer.calculate_field('code', 'Square')

# plot the data
ax = test_layer.plot(color='0.8', edgecolor='0.4')
test_layer.label('code')
ax.set_title('Number of rows: {}'.format(test_layer.count))
```

## Deleting data

Rows can be deleted from a table using the `delete_selected_rows()` method. As the name implies, the `delete_selected_rows()` method only deletes rows that are selected in the table, and can be used in conjuction with selection methods to select a set of rows to be deleted.

```python
# let's use a point to select one of the squares
# in our test dataset and delete it

point_in_right_square = Point({
    'x': 20,
    'y': 4
}, test_layer.spatial_reference)
test_layer.select_layer_by_location('CONTAINS', point_in_right_square)
test_layer.delete_selected_rows()

# plot the data
ax = test_layer.plot(color='0.8', edgecolor='0.4')
test_layer.label('code')
ax.set_title('Number of rows: {}'.format(test_layer.count))
```

## Performing an identity overlay

Identity overlays involve subdividing polygons in a layer based on the boundaries of geometries in another layer, and then marking the subdivided layer with new attributes derived from the cutting layer. In OSGIS processing this is the primary method for producing sublot cuts.

Geopandas geodataframes also have an `overlay()` method, but I have found cases where this method returns incorrect results so I recommend using the Models API method anytime an identity overlay is required.

```python
# create a set of zoning district sublots 
# using the `identity_to_layer()` method

# notice that this method creates a new layer 
# in the geodatabase called "zoning_district_sublots"
# which contains the outputs of the analysis

# also, specifying "ZONEDIST" in the `join_attributes` parameter
# will flag the subdivided outputs with values from the ZONEDIST
# field in the zoning districts table
zd_sublots_layer = example_lots_layer.identity_to_layer(example_zd_layer, 'zoning_district_sublots', join_attributes='ZONEDIST')
ax = zd_sublots_layer.plot(column='ZONEDIST', cmap='tab20c')
ax.set_title('Zoning district sublots')
```

## Dissolving layers

Another useful analysis method is `dissolve_to_layer()` which "dissolves" (i.e. merges together) adjacent polygons that have matching attributes. The output of the `dissolve_to_layer()` method is a new feature class with attributes matching the attributes provided to the dissolve function. 

```python
# dissolve the zd_sublots layer based on zoning district
# note the `multi_part` flag, which tells the function to
# create single part geometries
dissolved_layer = zd_sublots_layer.dissolve_to_layer('dissolved_zds', dissolve_fields='ZONEDIST', multi_part=False)

# plot the map and grab the axes its drawn in
ax = dissolved_layer.plot(column='ZONEDIST', cmap='tab20c')
ax.set_title('Zoning districts')

# label the plot based on the 'ZONEDIST' attribute
dissolved_layer.label('ZONEDIST')
```

## Creating complex selections

In some cases a layer may need to be queried in a very complex way, including making multiple selections with combinations of spatial and attribute queries. In the examples below, we'll build complex queries using the `selection_type` parameter of the `select_layer_by_attribute()` and `select_layer_by_location()` methods.

### Using subset selections and remove from selection

In the first example, we will select some features using a combination of spatial and attribute queries, and then filter the results further by removing items from the current selection.

```python
# returning to our example lots and zoning district layers, 
# let's create some complex spatial/attribute queries

# we'll go ahead and plot the lots layer
# and then plot the selected lots on top of it
# (make sure the selection set is clear on the lots layer before plotting)
example_lots_layer.clear_selection()
ax = example_lots_layer.plot(color='0.8', edgecolor='0.4')

# for our first example let's select lots that
# 1) spatially intersect the 'R8B' zoning districts and
# 2) are inside of or overlap a corner area (`inCorn` > 0)

# first, select the 'R8B' zoning district in the zoning districts layer
example_zd_layer.select_layer_by_attribute("ZONEDIST = 'R8B'")

# now spatially select the lots that intersect the 'R8B' zoning districts
example_lots_layer.select_layer_by_location('INTERSECT', example_zd_layer)

# finally, create a subset selection of lots that have `inCorn` > 0
example_lots_layer.select_layer_by_attribute("inCorn > 0", selection_type='SUBSET_SELECTION')

# plot the results
ax = example_lots_layer.plot(ax=ax, color='yellow', edgecolor='0.4')
title_text = 'First selection: {} lots selected'.format(example_lots_layer.count)

# for our next example, we'll remove the lots
# from this selection that intersect the 'R10' zoning district areas
# (on the east and west ends of the block)

# first, select the 'R10' zoning district in the zoning districts layer
example_zd_layer.select_layer_by_attribute("ZONEDIST = 'R10'")

# next, remove the lots that intersect the R10 zones from the selection set
example_lots_layer.select_layer_by_location('INTERSECT', example_zd_layer, selection_type='REMOVE_FROM_SELECTION')

# plot the results in our current drawing
ax = example_lots_layer.plot(ax=ax, color='red', edgecolor='0.4')
title_text = '{}\nSecond selection: {} lots selected'.format(title_text, example_lots_layer.count)
ax.set_title(title_text)
```

### Using switch selection

In some cases it is easier to define which features you do _not_ want in your selection set, select those, and then "switch" the selection to select the features you're interested in. In the example below we'll select some lots by attribute and then switch the selection to select the opposite set of features.

```python
# for this example, we'll select all of the lots in the lots layer
# that are in or near a corner area (`inCorn` > 0) and display them in yellow,
# and then switch the selection and display the remaining lots in red

# make sure that the selection set is clear from the
# previous example and grab out lot count for labeling
example_lots_layer.clear_selection()
title_text = 'Total lots: {}'.format(example_lots_layer.count)

# first, select the inCorn features in the lots layer
# and plot the results
example_lots_layer.select_layer_by_attribute('inCorn > 0')
ax = example_lots_layer.plot(color='yellow', edgecolor='0.4')
title_text = '{}\nTotal inCorn lots: {}'.format(title_text, example_lots_layer.count)

# now switch the selection to select the opposite set of lots
example_lots_layer.switch_selection()
ax = example_lots_layer.plot(ax=ax, color='red', edgecolor='0.4')
title_text = '{}\nTotal non-inCorn lots: {}'.format(title_text, example_lots_layer.count)
ax.set_title(title_text)

# clear the selection before moving to the next example
example_lots_layer.clear_selection()
```

## Geopandas integration

Being able to access data in a dataframe is useful for many data science projects. The Models API `Layer()` object can easily be exported to a Geopandas geodataframe object using the `get_geodataframe()` method.

```python
# convert the example lots layer to a geodataframe
lots_gdf = example_lots_layer.get_geodataframe()

# plot the geodataframe
ax = lots_gdf.plot(color='0.8', edgecolor='0.4')
ax.set_title('Plot from geodataframe')
```
