# CitiBike JSON Import Example

This example demonstrates how the Models API can be used to import a json file into a geodatabase table and give the table spatial attributes.

## Examing the inputs

The input data for this example represents the locations of CitiBike stations in NYC. The source json can be downloaded using the link below.

[CitiBikeNYC source data](https://gbfs.citibikenyc.com/gbfs/en/station_information.json)

A sample record from the CitiBike dataset is below. A few things to note about the structure of the data:

1. Some fields that are represented by `list` or `dict` data types. We will not import these fields into the dataset in this example.
2. The dataset does not contain formal spatial data, but does contain lat/lon values in mercator degrees. We can use these values to construct geometries for the geodatabase.

```json
{

	"has_kiosk":true,
	"eightd_has_key_dispenser":false,
	"short_name":"6926.01",
	"name":"W 52 St & 11 Ave",
	"lat":40.76727216,
	"external_id":"66db237e-0aca-11e7-82f6-3863bb44ef7c",
	"station_type":"classic",
	"eightd_station_services":[],
```

## Importing the required libraries

In this example we will need access to the `Workspace()`, `Layer()` and `Point()` objects from the Models API, as well as libraries for working with json and making url request.

````python
import json
import requests

from envelopegis.models.workspace import Workspace
from envelopegis.models.layer import Layer
from envelopegis.models.point import Point
````

## Set up a working geodatabase

It's a good practice to create a new geodatabase for processing data to insure that no unintended changes occur in the production dataset. In the code below we will create a new working geodatabase to import the json into.

````python
# create a working geodatabase called 'citibike_stations'
working_credentials = {
    'host': 'gisbox.ehq',
    'port': 7602,
    'user': 'postgres',
    'password': ''
}
workspace = Workspace.create_geodatabase(working_credentials, 'citibike_stations')
````

## Pulling the bike station data from the web

We will make a web request to retrieve the bike station data in json format from the CitiBikeNYC website.

````python
# use a GET request to get the citibike data from the internet
station_data_url = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"
response = requests.get(station_data_url)

# The data representing the stations is located beneath a wrapper called "data"
stations = response.json()["data"]["stations"]

# pull the first station from the list and use it to
# print the fields available in each row
first_station = stations[0]
print('Fields available in the CitiBikeNYC stations json')
for key, value in first_station.items():
    print(f'\t{key}: {type(value)}')
````

## Creating a table to store our data

We can use a workspace object to create a new layer to store the json as point data. When adding fields to the layer we can use the data types from the json object to help determine the proper data types for the database fields.

````python
# create a point feature class called "citibike_stations_unprojected"
# with spatial reference 4326 (a geographic projection for lat/lon data)
stations_layer = workspace.create_feature_class('citibike_stations_unprojected', 'POINT', 
                                                spatial_reference=4326)

# now let's loop through the fields in the json inputs and
# create columns in our table based on them
for key, value in first_station.items():
    # for now we'll skip dictionaries and lists
    if isinstance(value, list) or isinstance(value, dict):
        continue
    
    # create a field based on data type
    if isinstance(value, str):
        stations_layer.add_field(key, 'TEXT', field_length=200)
    elif isinstance(value, bool):
        stations_layer.add_field(key, 'BOOLEAN')
    elif isinstance(value, float):
        stations_layer.add_field(key, 'FLOAT')
    elif isinstance(value, int):
        stations_layer.add_field(key, 'LONG')
    
# now get a list of fields in our table and print them out
print(f'Fields created in the {stations_layer.table_name} layer')
for field in stations_layer.fields:
    print(f'\t{field.name} ({field.type})')
````

## Creating geometries and populating the table

Now that we have established a layer for storing our data, we can loop through the json records and construct rows to be inserted into the layer. We can also use the values from each record's `lat` and `lon` fields to construct a proper geometry to store in the layer's shape field.

````python
# loop through the station data and create rows
# for the new layer
new_rows = []
for station in stations:
    # collect the non-spatial attributes for the row
    row = {}
    for key, value in station.items():
        # if we did not create a field for this key, this is
        # a dict or list field and we should skip it
        if key not in stations_layer.field_names:
            continue
        
        # add the attribute to the row
        row[key] = value
    
    # now let's create a geometry for the row based on lat/lon
    row[stations_layer.shape] = Point({'x': row['lon'], 'y': row['lat']}, 
                                      spatial_reference=stations_layer.spatial_reference)
    
    # add the new row to our list
    new_rows.append(row)

# insert the rows into the table
stations_layer.add_rows(new_rows)

# print the number of rows added
print(f'Added {stations_layer.count} rows to {stations_layer.table_name}')
````

## Projecting the table to New York State Plane

All of Envelope's GIS datasets are in the New York State Plane spatial reference, and the CitiBike dataset will not interact properly with our other data until it is reprojected to the proper spatial reference. 

````python
# reproject the layer to New York State Plane (2263)
final_layer = stations_layer.reproject_to_layer('citibike_stations', 2263)

# report our results
print(f'Reprojected {final_layer.count} points to spatial reference {final_layer.spatial_reference}')
````
